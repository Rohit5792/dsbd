```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
#import dataset
df=pd.read_csv('iris.csv')
X = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
y = df['Species']  # Target labels (species)
#use info or describe function to get the more details
```

```python
df.head()
```

```python
print(X)
```

```python
#standardise the data
Scaler=StandardScaler()
X_scaled=Scaler.fit_transform(X)
```

```python
#apply pca
pca=PCA(n_components=2)
X_pca=pca.fit_transform(X_scaled)
```

```python
#create new data frame with pca
pca_df = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
pca_df['species'] =y
```

```python
#print pca data
print(pca_df.head())
```

```python

# Plot the data points in the new PCA space
# Use Seaborn's scatterplot to plot the PCA results
plt.figure(figsize=(8, 6))
sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='Species', palette='Set1')

# Add labels and title
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Iris Dataset')

# Show the plot
plt.show()

```
